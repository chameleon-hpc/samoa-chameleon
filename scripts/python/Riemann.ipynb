# General Riemann solver for the Shallow Water Equations

```python
%pylab inline
from sympy import *
from sympy.plotting import plot, plot_parametric, plot3d, plot3d_parametric_surface
```

## General problem formulation

The general solution to the Riemann Problem for the 1D Shallow Water Equations consists of two waves emerging from the discontinuity, either one could be a rarefaction or a shock wave. We define:
$$
\Phi_{rare}(h_1, h_2) := 2 (\sqrt{g h_1} - \sqrt{g h_2})\\
\Phi_{shock}(h_1, h_2) := (h_1 - h_2) \sqrt{\frac{g}{2} (1 / h_1 + 1 / h_2)}
$$

and set for $h_l, h_r > 0$ and $u_l, u_r \in \mathbb R$:
$$
\Phi_l(h) := u_l + \left\{ \begin{array}{ll}\Phi_{rare}(h_l, h) & \text{if } h < h_l \\ \Phi_{shock}(h_l, h) & \text{otherwise} \end{array} \right. \\
\Phi_r(h) := u_r + \left\{ \begin{array}{ll}\Phi_{shock}(h, h_r) & \text{if } h > h_r \\ \Phi_{rare}(h, h_r) & \text{otherwise} \end{array} \right. \\
$$

Then the following condition must hold for the water height $h_m$ of the intermediate state:
$$
\Phi_l(h_m) = \Phi_r(h_m)
$$
where $u_m := \Phi_l(h_m) = \Phi_r(h_m)$ is the velocity of the intermediate state.


```python
g,hl,hr,h = symbols('g, hl, hr, h', real=True, positive=True)
ul,ur,u = symbols('ul, ur, u', real=True)

phi_rare = lambda h1, h2 : 2.0 * (sqrt(g * h1) - sqrt(g * h2))
phi_shock = lambda h1, h2 : (h1 - h2) * sqrt(g / 2.0 * (1.0 / h1 + 1.0 / h2))

phi_l = ul + Piecewise((phi_rare(hl, h), h < hl), (phi_shock(hl, h), True))
phi_r = ur + Piecewise((phi_shock(h, hr), h > hr), (phi_rare(h, hr), True))

pprint("Solve for h:")
pprint(Eq(phi_l, phi_r))
```

## Intermediate state solution

In order to solve the equation $\Phi_l(h) = \Phi_r(h)$, we use a nonlinear solver that finds a root of the function
$\Phi := \Phi_r - \Phi_l$.

```python
import pylab as pl

def compute_intermediate_state(hl_test, hr_test, ul_test, ur_test, g_test):
    g,hl,hr,h,t = symbols('g, hl, hr, h, t', real=True, positive=True)
    ul,ur,u,x = symbols('ul, ur, u, x', real=True)

    if (hr_test == 0.0):
        u_h = phi_l.subs({hl:hl_test, hr:hr_test, ul:ul_test, g:g_test})
        hm = 0.0
    elif (hl_test == 0.0):
        u_h = phi_r.subs({hl:hl_test, hr:hr_test, ul:ul_test, g:g_test})
        hm = 0.0
    else:
        u_h = phi_l.subs({hl:hl_test, hr:hr_test, ul:ul_test, g:g_test})
        phi = simplify(phi_r - phi_l).subs({hl:hl_test, hr:hr_test, ul:ul_test, ur:ur_test, g:g_test})
    
        hm = nsolve(phi, h, 0.5 * (hl_test + hr_test))

    um = u_h.evalf(subs={h:hm})
    
    return hm, um
```

## Analytic solution from intermediate state

We can now compute the solution from the intermediate state $(h_m, u_m)$, knowing that the first wave is a rarefaction when $h_l > h_m$ and a shock when $h_l < h_m$. The second wave is a shock when $h_m > h_r$ and a rarefaction when $h_m < h_r$.

```python
def compute_solution(hl, ul, hm, um, hr, ur, g):
    t = symbols('t', real=True, positive=True)
    x = symbols('x', real=True)
    
    s1 = 0.0
    s2 = 0.0

    if hl > hm:
        # 1-rarefaction:
        s1l = ul - sqrt(g * hl)
        s1 = um - sqrt(g * hm)
        h1 = 1.0/(9.0 * g) * (ul + 2.0 * sqrt(g * hl) - x / t) ** 2.0
        u1 = ul + 2.0 * (sqrt(g * hl) - sqrt(g * h1))
    elif hl < hm:
        # 1-shock:
        s1l = (hm * um - hl * ul) / (hm - hl)
        s1 = s1l
        h1 = hl
        u1 = ul
        
    if hm > hr:
        # 2-shock:
        s2r = (hm * um - hr * ur) / (hm - hr)
        s2 = s2r
        h2 = hr
        u2 = ur
    elif hm < hr:
        # 2-rarefaction:
        s2r = ur + sqrt(g * hr)
        s2 = um + sqrt(g * hm)
        h2 = 1.0/(9.0 * g) * (ur - 2.0 * sqrt(g * hr) - x / t) ** 2.0
        u2 = ur - 2.0 * (sqrt(g * hr) - sqrt(g * h2))
        
    if (hm == hl):
        s1l = s2
        s1 = s2
        h1 = hl
        u1 = ul    
        
    if (hm == hr):
        s2r = s1
        s2 = s1
        h2 = hr
        u2 = ur
        
    return Piecewise((hl, x < s1l * t), (h1, x < s1 * t), (hm, x < s2 * t), (h2, x < s2r * t), (hr, True)), \
           Piecewise((ul, x < s1l * t), (u1, x < s1 * t), (um, x < s2 * t), (u2, x < s2r * t), (ur, True))
```

```python
def Riemann(hl_test, hr_test, ul_test, ur_test, g_test):
    hm, um = compute_intermediate_state(hl_test, hr_test, ul_test, ur_test, g_test)
    h_sol, u_sol = compute_solution(hl_test, ul_test, hm, um, hr_test, ur_test, g_test)
    
    print "Test   : hl =", hl_test, "hr =", hr_test, " ul =", ul_test, " ur =", ur_test, " g =", g_test
    print "Result : hm =", hm, " um =", um, " hum =", hm * um
    
    h = symbols('h', real=True, positive=True)
    u = symbols('u', real=True)

    pprint(Eq(h, h_sol))
    animate(h_sol)
    
    pprint(Eq(u, u_sol))
    animate(u_sol)
    
    return h_sol, u_sol

def animate(f):
    t = symbols('t', real=True, positive=True)
    x = symbols('x', real=True)

    X_test = np.linspace(-5.0, 5.0, 256, endpoint=True)
    f_xt = lambda x_test, t_test : f.evalf(subs={x:x_test, t:t_test})
   
    for t_test in np.linspace(0.5, 2.0, 4, endpoint=True):
        pl.plot(X_test, [f_xt(x_test, t_test) for x_test in X_test])
    pl.show()
```
